```python
tablero_inicial= ['   |   |   ',
                  ' 7 | 8 | 9 ',
                  '   |   |   ',
                  '-----------',
                  '   |   |   ',
                  ' 4 | 5 | 6 ',
                  '   |   |   ',
                  '-----------',
                  '   |   |   ',
                  ' 1 | 2 | 3 ',
                  '   |   |   ']

alternador = {1:1, -1:2}

def inicializar():
    #Regresamos todos los valores a los originales para volver a jugar
    global simbolo, tablero, matriz 
    simbolo = {'jugador0':' ',
               'jugador1':' ',
               'jugador2':' '}

    tablero = [list(linea) for linea in tablero_inicial]
    for i, linea in enumerate(tablero):
        for j, letra in enumerate(linea):
            if letra in [x for x in range(10)]:
                tablero[i][j] = ' '

    matriz = [[0,0,0],[0,0,0],[0,0,0]]

def cambia_tablero():
    #Cambiamos el tablero de acuerdo al estado de la matriz
    for i in range(3):
        for j in range(3):
            tablero[1+i*4][1+j*4] = simbolo['jugador'+str(matriz[i][j])]

def imprime():
    #Imprimimos el tablero actual
    for linea in tablero:
        for letra in linea:
            print(letra, end = "")
        print()
        
def ganador():
    #Checa si alguno de los dos jugadores ya ganó
    #OUTPUT: True o False
    
    for fila in matriz:
        if 0 not in fila:
            if (1 not in fila) or (2 not in fila):
                return True
    
    for i in range(3):
        columna = [matriz[j][i] for j in range(3)]
        if 0 not in columna:
            if (1 not in columna) or (2 not in columna):
                return True
    
    diagonal = [matriz[i][i] for i in range(3)]
    if 0 not in diagonal:
        if (1 not in diagonal) or (2 not in diagonal):
            return True
        
    diagonal = [matriz[2-i][i] for i in range(3)]
    if 0 not in diagonal:
        if (1 not in diagonal) or (2 not in diagonal):
            return True
    
    return False

def anota(jugador, numero):
    #Recibe un numero del 1 al 9 de algun jugador, y modifica la matriz de estado
    #La correspondencia de la posición es como en el numpad
    numero-=1
    matriz[2-numero//3][numero%3] = jugador
    
def pregunta(string, valid, explica = ''):
    #Pregunta algo hasta obtener una respuesta valida
    ans = input(string)
    while ans not in valid:
        print("Valor no valido. "+explica)
        ans = input(string)
    return ans
```

```python
print('Hola, bienvenido al juego del gato')

juega = True
while juega:
    inicializar()

    simbolo['jugador1'] = pregunta('Jugador1: ¿Qué simbolo quieres ser X ó O?  ',['X', 'O'])
    if simbolo['jugador1'] == 'X':
        simbolo['jugador2'] = 'O'
    else:
        simbolo['jugador2'] = 'X'

    comienza = ''
    first = 0
    comienza = pregunta("Jugador1: ¿Quieres comenzar tú, jugador1? (si/no)", 
                        ['si', 'no'], "Responde si o no")
    if comienza == "si":
        first = 1
    else:
        first = -1

    print(f"Bien, comienza el Jugador{alternador[first]}")

    imprime()
    first *=- 1
    contador = 0

    while not ganador():
        first *=-1
        numero = pregunta(f'Jugador{alternador[first]}: ¿Donde quieres poner? ',
                         '123456789', 'Elige una posición del 1 al 9 de las disponibles')
        anota(alternador[first], int(numero))
        cambia_tablero()
        imprime()
        contador += 1
        if contador >=9:
            print('Nadie gana, ¡Empate!')
            break
    else: 
        print(f'Felicidades jugador{alternador[first]}, ¡Ganaste!')
        
    continua = pregunta('¿Quieren jugar de nuevo? (si/no)  ', ['si', 'no'])
    if continua == 'no':
        juega = False

else:
    print('¡Hasta la proxima! :)')
```
